```python
independent = pd.DataFrame(train.drop(["Habitability_score"],axis=1))
dependent = pd.DataFrame(train["Habitability_score"])
```

```python
X,x,Y,y = train_test_split(independent,dependent,test_size=0.25,random_state=200)
```

from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
sc.fit(X)
X = sc.transform(X)
x = sc.transform(x)
sc.fit(Y)
Y = sc.transform(Y)
y = sc.transform(y)

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor(random_state=25, n_jobs=-1, max_depth=20,n_estimators=900, oob_score=True)
RF.fit(X,Y)
RF.score(x,y)
```

```python
pred = pd.DataFrame(RF.predict(test))
pred["Habitability_score"]=pred[0]
pred.drop([0],axis=1,inplace =True)
pred1 = pd.concat([var,pred],axis=1)
pred1.to_csv("predictions.csv", index = False)
```
